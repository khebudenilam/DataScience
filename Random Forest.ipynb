```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error,accuracy_score
```

```python
A=pd.read_csv("E:/Python/Dataset/Cars93.csv")
```

```python
A.head()
```

```python
Y=A[["Weight"]]
X=A.drop(labels=["Weight","Unnamed: 0","Make","Model"],axis=1)
```

```python
le=LabelEncoder()
X=X.apply(le.fit_transform)
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=30)
rfr=RandomForestRegressor(criterion="mae",random_state=30)
model=rfr.fit(xtrain,ytrain)
pred=model.predict(xtest)
print("MAE is %.3f"%mean_absolute_error(ytest,pred))
```

```python
rfr=RandomForestRegressor(criterion="mae",random_state=30)
#n_estimators=number of tree
#max_depth=cut the trees from max_depth value

tp={"n_estimators":range(1,50,1),"max_depth":range(2,20,1)}
cv=GridSearchCV(rfr,tp,scoring="neg_mean_absolute_error",cv=4)
cvmodel=cv.fit(xtrain,ytrain)
ntress=cvmodel.best_params_["n_estimators"]
md=cvmodel.best_params_["max_depth"]
rfr=RandomForestRegressor(criterion="mae",random_state=30,max_depth=md,n_estimators=ntress) 
model=rfr.fit(xtrain,ytrain)
pred=model.predict(xtest)
print("MAE is %.3f"%mean_absolute_error(ytest,pred))
    
```
