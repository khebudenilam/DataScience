```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
A=pd.read_csv("E:/Python/Dataset/iris.csv")
```

```python
A.head(2)
```

```python
Y=A[["Species"]]
X=A.drop(labels=["Unnamed: 0","Species"],axis=1)
```

```python
A.info()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
rfc = RandomForestClassifier(n_estimators=30,max_depth=3,random_state=30)
model=rfc.fit(xtrain,ytrain)
pred=model.predict(xtest)
print("accuracy score %.3f"%accuracy_score(ytest,pred))
```

```python
from sklearn.model_selection import  GridSearchCV
tp={"n_estimators":range(2,40,1),"max_depth":range(2,10,1)}
rfc=RandomForestClassifier(random_state=30)
cv=GridSearchCV(rfc,tp,scoring="accuracy",cv=4)
cvmodel=cv.fit(xtrain,ytrain)
cvmodel.best_params_
```
